```python
import pandas as pd
```

```python
data_trimmed = pd.read_csv("Food_Availability/trimmed_data/df_fluid_milk_trimmed.csv")
data_cleaned = pd.read_csv("Food_Availability/cleaned_data/df_fluid_milk.csv")
```

```python
def prepare_data_report(data):
    data_types = pd.DataFrame(
        data.dtypes,
        columns=['Data Type']
    )

    missing_data = pd.DataFrame(
        data.isnull().sum(),
        columns=['Missing Values']
    )

    unique_values = pd.DataFrame(
        columns=['Unique Values']
    )
    for row in list(data.columns.values):
        unique_values.loc[row] = [data[row].nunique()]

    maximum_values = pd.DataFrame(
        columns=['Maximum Value']
    )
    for row in list(data.columns.values):
        maximum_values.loc[row] = [data[row].max()]
    dq_report = data_types.join(missing_data).join(unique_values).join(maximum_values)
    dq_report["Col_Name"] = dq_report.index
    dq_report.reset_index(drop = True, inplace=True)
    dq_report_cols = list(dq_report.columns)
    new_cols = [dq_report_cols[-1]] + dq_report_cols[0:-1]
    dq_report = dq_report[new_cols]
    return dq_report
```

```python
dq_report_trimmed = prepare_data_report(data_trimmed)
dq_report_cleaned = prepare_data_report(data_cleaned)
```

```python
dq_report_trimmed.to_csv("Food_Availability/dq_report_trimmed.csv")
dq_report_cleaned.to_csv("Food_Availability/dq_report_cleaned.csv")
```

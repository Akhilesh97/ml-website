## Removing inconsistent values
1. Since we know that all the columns have floating point values, we can replace inconsistent values with the following logic - 
    - Try and return the type casted value of the record as a float,
    - If an exception is raised return the value as NaN and print the value. 

2. This way, if the type casted value is a floating point entered as a string, the value is returned correctly, else the inconsistent values is returned. 

3. The following steps can be performed for columns whose dtype is "object" 

### Before Removing Inconsistent Values

```python
df_dairy_trimmed.dtypes
```

```python
def replace_string_vals(x):
    try:
        return float(x)
    except Exception as e:
        print(e, x)
        return np.nan

def handle_inconsistent_vals(df):
    for col in df.columns:    
        if df[col].dtype == object:
            df[col] = df[col].apply(lambda x: replace_string_vals(x))
    return df
    
```

```python

for col in df_dairy_trimmed.columns:
    
    if df_dairy_trimmed[col].dtype == object:
        df_dairy_trimmed[col] = df_dairy_trimmed[col].apply(lambda x: replace_string_vals(x))
```

## Handle Missing Values
1. In most cases in this dataset, if there are missing values, means that there was no data availabe during that year.
2. It is also observed that most of the missing values are appear contigously before or after a given time period. For example - For the category of dairy - Fluid Cream, we can see that there was no data recorded after 2006. 
3. Hence, it would be fitting to replace these values with a 0 and ignore these values during visualization or model building. 

```python
def handle_missing_vals(df):
    df_ = df.fillna(0)
    return df_
```

```python
class data_cleaning:
    
    def __init__(self, df, filename, out_filepath):
        self.df = df
        self.filename = filename
        self.out_filepath = out_filepath
        
    def check_duplicates(self):
        print("Number of Duplicate rows")
        print("*"*100)
        print(len(self.df[self.df.duplicated()]))
        if len(self.df[self.df.duplicated()]) > 0:
            self.df = self.df.drop_duplicates()
        
    def replace_string_vals(self, x):
        try:
            return float(x)
        except Exception as e:
            print(e, x)
            return np.nan

    def handle_inconsistent_vals(self):
        for col in self.df.columns:    
            try:
                if self.df[col].dtype == object:
                    self.df[col] = self.df[col].apply(lambda x: self.replace_string_vals(x))
            except Exception as e:
                print(self.df[col])
    
    def handle_missing_vals(self):
        self.df = self.df.fillna(0)
    
    def clean_data(self):
        self.check_duplicates()
        self.handle_inconsistent_vals()
        self.handle_missing_vals()
        self.df.to_csv("%s/%s.csv"%(self.out_filepath, filename))
    
```

```python
import os
```

```python
for file in os.listdir("Food_Availability/trimmed_data/"):
    if file.endswith(".csv"):
        print("Processing file...%s"%file)
        df_trimmed = pd.read_csv("Food_Availability/trimmed_data/%s"%file)     
        new_cols = []
        for col in df_trimmed:
            col_ = ''.join([i for i in col if not i.isdigit()])
            new_cols.append(col_)
        df_trimmed.columns = new_cols
        print()
        filename = file.replace("_trimmed.csv","")
        out_filepath = "Food_Availability/cleaned_data/"
        obj = data_cleaning(df_trimmed, filename, out_filepath)
        obj.clean_data()                
        print("Done Processing")
        print("-"*100)
```

## Fruits

```python
fruits_excel = pd.ExcelFile("Food_Availability/raw_data/frtot.xls")

fruits_excel.close()

fruits_excel_sheets = fruits_excel.sheet_names[2:]

fruits_dict = {}
for sheet_name_ in fruits_excel_sheets:
    df_fruits = pd.read_excel("Food_Availability/raw_data/frtot.xls", sheet_name=sheet_name_)
    df_fruits_cols = list(df_fruits.iloc[0].values)
    df_fruits_trimmed = df_fruits.iloc[3:-3,:]
    df_fruits_trimmed.columns = df_fruits_cols
    fruits_dict[sheet_name_] = df_fruits_trimmed

fruits_dict.keys()

for fruit in fruits_dict.keys():
    df_to_write = fruits_dict[fruit]
    df_to_write.to_csv("Food_Availability/trimmed_data/fruits/df_%s_trimmed.csv"%fruit)
```
